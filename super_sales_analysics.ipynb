# Data load 

```python
import requests
# raw csv file ka URL
url = "https://raw.githubusercontent.com/WuCandice/Superstore-Sales-Analysis/refs/heads/main/dataset/Superstore%20Dataset.csv"
# Request bhejna
response = requests.get(url)
# Check karo request successful hai ya nahi
if response.status_code == 200:
    with open("superstore.csv", "w") as file:
        file.write(response.text)
    print("CSV file successfully download ho gayi!")
else:
    print("Error:", response.status_code)
```

# 🧹 3. Data Cleaning (date convert, missing check)

```python
# Date columns ko convert karna
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Ship Date'] = pd.to_datetime(df['Ship Date'])

# Missing values check
print(df.isnull().sum())
# Explanation:
# Dates ko datetime format mein convert kiya.
# Missing values check kiya.
```

# 📅 4. New Columns banana (Month, Year)

```python

df["Year"] = df["Order Date"].dt.year


df["Month"] = df["Order Date"].dt.month
# Explanation:
# Month aur Year nikal kar time-based analysis ke liye ready kiya.
```

```python
df.info()
```

# 📊 5. Basic Analysis (sales, profit)

```python
# Total sales aur profit
print("Total Sales:", df['Sales'].sum())
print("Total Profit:", df['Profit'].sum())
# Explanation:
# Business ka overall performance samajhne ke liye total sales aur profit calculate kiya.
```

 # 🧑‍🤝‍🧑 6. GroupBy Analysis (category, region)

```python
# Category wise sales aur profit
print(df.groupby('Category')[['Sales','Profit']].sum())

# Region wise sales aur profit
print(df.groupby('Region')[['Sales','Profit']].sum())
# Explanation:
# Category aur Region ke hisaab se sales aur profit ka breakdown.
```

# 📉 7. Loss Analysis

```python
# Negative profit wale rows
loss_data = df[df['Profit'] < 0]
print(loss_data[['Product Name','Sales','Profit']].head())
# Explanation:
# Negative profit wale products identify kiye jo business ko nuksan de rahe hain.
```

# 🏆 8. Top/Bottom Products

```python
# Top 10 products by profit
top_products = df.groupby('Product Name')['Profit'].sum().sort_values(ascending=False).head(10)
print(top_products)

# Bottom 10 products by profit
bottom_products = df.groupby('Product Name')['Profit'].sum().sort_values().head(10)
print(bottom_products)
# Explanation:
# Top 10 aur Bottom 10 products nikal kar performance samjha.
```

# 📈 9. Visualization (graphs)

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Sales trend by month
plt.figure(figsize=(10,5))
sns.lineplot(data=df.groupby('Month')['Sales'].sum())
plt.title("Monthly Sales Trend")
plt.show()

# Category wise profit
plt.figure(figsize=(8,5))
sns.barplot(x='Category', y='Profit', data=df)
plt.title("Profit by Category")
plt.show()
# Explanation:
# Graphs se trends aur patterns easily samajh aate hain.
```

# 📊 Extra Visualizations

```python
# 📌 Region-wise Sales Pie Chart
plt.figure(figsize=(7,7))
df.groupby('Region')['Sales'].sum().plot(kind='pie', autopct='%1.1f%%')
plt.title("Region-wise Sales Distribution")
plt.ylabel("")
plt.show()

# 📌 Yearly Sales Trend
plt.figure(figsize=(10,5))
sns.lineplot(x='Year', y='Sales', data=df, estimator='sum')
plt.title("Yearly Sales Trend")
plt.show()

# 📌 Discount vs Profit Scatter Plot
plt.figure(figsize=(8,5))
sns.scatterplot(x='Discount', y='Profit', data=df, alpha=0.6)
plt.title("Discount vs Profit Relationship")
plt.show()

# 📌 Sub-Category wise Profit
plt.figure(figsize=(12,6))
sns.barplot(x='Sub-Category', y='Profit', data=df, estimator='sum')
plt.xticks(rotation=45)
plt.title("Profit by Sub-Category")
plt.show()
```
